# Apa itu kuantum?

## Apa itu bit?

Pada bab sebelumnya kita telah mempelajari bahwa 'bits' adalah alfabet paling sederhana di dunia, dan mereka dapat digunakan untuk mewakili informasi apa pun. Kita juga belajar bahwa semua komputer modern menyimpan, dan melakukan operasi pada bits. Sejauh ini, kita hanya memikirkan bits sebagai unit informasi abstrak, tetapi untuk membuat komputer yang berkerja dengan benar, kita perlu membuat bits dari hal-hal yang nyata. Jadi, bagaimana kita bisa membuat bits di dunia nyata?

<!-- ::: q-block.tabs -->

## Bagaimana kita bisa menyimpan bits?

<!-- ::: tab -->

### Kartu berlubang

<!-- ::: column(width=350) -->

**Kartu berlubang**

Pada masa awal-awal komputasi, ilmuwan komputer menyimpan bit dengan membuat lubang di kartu kertas. Kartu-kartu ini dibagi menjadi kisi-kisi, dan setiap sel dalam kisi-kisi itu mewakili bit. Jika ada lubang di sel itu, bitnya adalah '1', jika tidak ada lubang, bitnya adalah '0'. Jika Anda ingin mengubah nilai bit, Anda dapat membuat lubang baru, atau [menambal](gloss:patch) lubang itu kembali.

<!-- ::: column(width=250) -->

![gambar kartu berlubang](images/what-is/punched-card.png)

<!-- ::: -->

<!-- ::: tab -->

### Cakram Padat (Compact disks "CD")

<!-- ::: column(width=350) -->

**Cakram Padat**

Cakram padat dibuat pada tahun 80-an untuk menyimpan rekaman audio menggunakan bits. Perancang memutuskan untuk menggunakan lubang dan palung pada permukaan disk untuk mewakili bit mereka, dan kita dapat menggunakan laser di atas disk untuk mendeteksinya. Disk dibaca sepanjang garis spiral, dibagi menjadi beberapa segmen. Untuk setiap segmen, permukaan datar mewakili '0', dan transisi dari puncak ke lembah (atau sebaliknya) mewakili '1'.

<!-- ::: column(width=250) -->

![Ilustrasi CD dengan zoom-in yang menggambarkan pola alur](images/what-is/cd.png)

<!-- ::: -->

<!-- ::: tab -->

### Orbital elektron

<!-- ::: column(width=350) -->

**Orbital elektron**

Anda mungkin ingat dari pelajaran kimia (jangan khawatir jika tidak) bahwa partikel yang disebut elektron mengorbit pusat (inti) atom. Anda mungkin juga ingat bahwa elektron-elektron ini hanya dapat mengorbit pada jarak tertentu ( [terkuantisasi](gloss:quantized) ) dari inti yang kita sebut [*kulit*](gloss:shell) . Jika kita memiliki cara untuk mendeteksi di kulit mana elektron berada, kita dapat memilih dua dari kulit ini dan menggunakan lokasi elektron sebagai bit. Jika elektron berada di salah satu kulit ini, bitnya adalah '0', dan jika berada di kulit yang lain, bitnya adalah '1'.

<!-- ::: column(width=250) -->

![Ilustrasi yang menunjukkan atom sederhana dengan satu elektron dan dua kulit](images/what-is/atom.png)

Perilaku kartu berlubang dan compact disk dapat dengan mudah dijelaskan menggunakan *fisika klasik*: Fisika pada dasarnya adalah deskripsi dunia di sekitar kita; kita melakukan eksperimen untuk melihat bagaimana dunia berperilaku, dan dari sini kita mencoba mencari tahu aturan apa yang diikuti alam semesta. "Fisika klasik" adalah nama untuk seperangkat aturan yang dibuat oleh para ilmuwan sebelum awal 1900-an, dan dengan sangat baik dapat memprediksi perilaku hal-hal seperti bola kriket dan mesin mobil.

Tetapi elektron yang mengorbit sebuah atom sedikit berbeda. Sekitar awal 1900-an, para ilmuwan mulai dapat mempelajari berbagai hal dalam skala atom. Mereka menemukan bahwa hal-hal yang sangat kecil seperti atom berperilaku berbeda dengan hal-hal yang berinteraksi dengan kita dalam kehidupan kita sehari-hari, dan dalam kasus-kasus tertentu aturan fisika klasik tidak sepenuhnya benar. Fisika yang mereka butuhkan perlu untuk dimodifikasi, sehingga para ilmuwan datang dengan seperangkat aturan yang lebih akurat yang dikenal sebagai "fisika kuantum".

### Eksperimen kunci dalam fisika kuantum

Bagi mereka yang tertarik, berikut adalah ringkasan singkat dari beberapa peristiwa yang mengarah pada perkembangan fisika kuantum. Anda tidak perlu memahami ini untuk membaca sisa kursus ini, ini ditujukan bagi mereka yang ingin membaca lebih lanjut tentang sejarah dan fisika.

<summary>Bencana ultraviolet</summary><p> Benda panas memancarkan radiasi sebagai cahaya, dan semakin tinggi suhunya, semakin tinggi frekuensi gelombang cahaya ini. Jika sesuatu hanya memancarkan dan meradiasikan cahaya (yaitu, tidak memantulkannya), kita menyebutnya 'benda hitam'. Benda hitam yang secara teoritis sempurna bukanlah perkiraan yang mengerikan dari banyak objek nyata, termasuk bintang dan planet.</p>
<p> Antara 1900-1905, dua ilmuwan (Rayleigh dan Jeans) menggunakan aturan fisika klasik untuk mendapatkan persamaan yang menjelaskan bagaimana <i>suhu</i> benda hitam mempengaruhi <i>frekuensi</i> gelombang cahaya yang dipancarkannya. Hukum yang mereka peroleh meramalkan bahwa benda panas harus memancarkan cahaya frekuensi tinggi (ultraviolet) dalam jumlah yang tidak masuk akal, yang tidak sesuai dengan apa yang kita lihat sama sekali.</p>
<p> Sampai saat ini, diterima secara luas bahwa cahaya adalah gelombang. Tetapi ilmuwan lain (Max Planck) berhasil menunjukkan bahwa, jika cahaya harus dipancarkan dalam potongan terkuantisasi (yaitu, partikel yang sekarang kita sebut 'foton'), Anda dapat memperoleh persamaan yang menggambarkan dengan tepat hubungan antara suhu dan emisi yang dapat kita peroleh ukuran.</p>

<details> <summary>Percobaan celah ganda</summary><p> Jika Anda menembakkan beberapa partikel melalui lubang ke layar dan merekam di mana mereka mendarat, Anda akan menemukan lebih banyak di tengah layar dan lebih sedikit di tepinya (bentuknya sebenarnya sedikit lebih rumit dari ini, tetapi kita tidak akan membahasnya itu disini). Jadi, apa yang Anda harapkan terjadi jika Anda menembakkan partikel melalui _dua_ lubang ke layar? Jika lubangnya cukup dekat, akal sehat mengatakan kita akan melihat hal yang kira-kira sama, dengan mungkin lebih banyak partikel yang melewatinya.</p>
<p> Tetapi ketika kita mencoba eksperimen ini dengan partikel kecil seperti foton, elektron, dan bahkan atom, kita melihat pola interferensi, bahkan jika kita hanya mengirim satu partikel pada satu waktu! Eksperimen ini dilakukan pada akhir 1700-an dengan cahaya, yang membuat para ilmuwan percaya bahwa cahaya bukanlah partikel melainkan gelombang. Tapi seperti yang kita lihat dengan bencana ultraviolet, ini masih belum keseluruhan dari apa yang sebenarnya terjadi.</p> ![](images/what-is/Slit-slit-and-double-slit.jpg)<p> Gambar oleh Jordgette - Karya sendiri, CC BY-SA 3.0, <a href="https://commons.wikimedia.org/w/index.php?curid=9529698">Tautan</a></p></details>

Untuk ilmuwan komputer, pertanyaan yang muncul adalah: "Bagaimana jika bit kita mengikuti aturan fisika kuantum, bukan fisika klasik?". Kita menyebut bit ini "qubit" untuk "bit kuantum", dan komputer yang dapat beroperasi pada bit ini disebut "komputer kuantum".

Di bagian berikutnya Anda dapat berinteraksi dengan salah satu qubit ini (sebenarnya, tidak juga – kita hanya mensimulasikannya - tetapi kami berjanji bahwa perilakunya sama) dan temukan apa itu 'kuantum'!

## Mengeksplorasi qubits

Untuk mengeksplorasi perilaku kuantum, kita perlu mengingatkan diri kita sendiri tentang probabilitas. Jika Anda telah membaca bab sebelumnya, semua operasi yang telah kita lihat sejauh ini bersifat [deterministik](gloss:deterministic). Ini berarti bahwa, bekerja pada keadaan input yang sama, mereka akan selalu memberikan keadaan output yang sama. Mari kita lihat operasi baru yang hanya bekerja pada bit kuantum: gerbang [Hadamard](gloss:hadamard), yang akan kita sebut "gerbang-H" untuk singkatan.

```
q-mini-composer(goal="what-is-minicomposer")
    .slide
        .circuit
            .autoMeasureGate
            .availableGates H
            .initialCircuit
                .qubit
            .goalCircuit
                .qubit H
            .startProbabilities 0: 1, 1: 0
            .endProbabilities 0: 0.5, 1: 0.5
        .instructions Drag the H-gate down onto the circuit
        .lesson The H-gate seems to give a 50-50 chance of transforming the qubit from |0⟩ to either |0⟩ or |1⟩. We can plot this on a probability tree: <img src="images/what-is/minicomposer/prob-0.svg">
        .info Nice one! Let's keep experimenting.
    .slide
        .circuit
            .autoMeasureGate
            .availableGates X
            .initialCircuit
                .qubit
            .goalCircuit
                .qubit X
            .startProbabilities 0: 1, 1: 0
            .endProbabilities 0: 0, 1: 1
        .instructions Transform our qubit to the state |1⟩.
        .lesson Great! After the X-gate, our qubit has 100% chance of being in the state |1⟩. Let's see how the H-gate acts on this input.
        .info Click 'next' to add the H-gate.
    .slide
        .circuit
            .autoMeasureGate
            .availableGates H
            .initialCircuit
                .qubit X
            .goalCircuit
                .qubit X H
            .startProbabilities 0: 0, 1: 1
            .endProbabilities 0: 0.5, 1: 0.5
        .instructions Add the H-gate to the end of the circuit.
        .lesson The H-gate also seems to give a 50-50 chance of transforming the qubit from |1⟩ to either |0⟩ or |1⟩. We can plot this on a probability tree too: <img src="images/what-is/minicomposer/prob-1.svg">
        .info Awesome!
    .slide
        .circuit
            .autoMeasureGate
            .availableGates
            .initialCircuit
                .qubit X H
            .goalCircuit
                .qubit H
            .startProbabilities 0: 0.5, 1: 0.5
            .endProbabilities 0: 0.5, 1: 0.5
        .instructions Remove the X-gate from your circuit
        .lesson Now we know how the H-gate behaves on both possible input states, we can use this to predict its behaviour. <img src="images/what-is/minicomposer/prob-chained.svg"> Look at the probability tree, what do you think would happen if we applied two H-gates in a row?
        .info Think about the answer before clicking 'next'.
    .slide
        .circuit
            .autoMeasureGate
            .availableGates H
            .initialCircuit
                .qubit H
            .goalCircuit
                .qubit H H
            .startProbabilities 0: 0.5, 1: 0.5
            .endProbabilities 0: 1, 1: 0
        .instructions Let's test our theory. What would happen if we applied <i>two</i> H-gates in sequence?
        .lesson What‽ This is strange. Let's try with the input |1⟩.
        .info Nice one! Let's keep experimenting.
    .slide
        .circuit
            .autoMeasureGate
            .availableGates H
            .initialCircuit
                .qubit X H
            .goalCircuit
                .qubit X H H
            .startProbabilities 0: 0.5, 1: 0.5
            .endProbabilities 0: 0, 1: 1
        .instructions What happens if we apply two H-gates after an X-gate?
        .lesson This doesn't fit our model at all! In fact, there are no numbers we can put on our probability trees that will describe this behaviour. <img src="images/what-is/minicomposer/prob-chained-wrong.svg"> To explain this, we'll need to replace our probabilities with something else.
        .info Congratulations! You've completed this exercise.
```

Kita baru saja melihat beberapa perilaku yang menarik: Dua operasi probabilistik, diterapkan secara berurutan, tampaknya "membatalkan" satu sama lain! Faktanya, perilaku sederhana ini sangat tidak biasa sehingga kita tidak dapat menggambarkannya menggunakan pohon probabilitas. Tidak ada angka yang bisa kita letakkan di cabang kita yang memberikan hasil yang benar dalam semua kasus.

Anda mungkin berpikir, "mungkin gerbang-H tahu itu diterapkan dua kali, dan berperilaku berbeda?". Ini akan dihitung sebagai input lain pada gerbang, di samping status qubit. Anda dapat menjelaskan perilaku ini jika Anda mengizinkan input tersembunyi lainnya ke operasi gerbang-H yang tidak dapat kita lihat secara langsung. Dan faktanya, inilah yang terjadi.

## Lebih dari probabilitas

Probabilitas berguna ketika ada banyak kemungkinan hasil dan kita tidak memiliki cukup informasi untuk menentukan mana yang akan terjadi, seperti lemparan dadu, atau lemparan koin. Kita memberikan setiap hasil probabilitas dan menggunakan ini untuk mengetahui kemungkinan sesuatu terjadi. Probabilitas bekerja sangat baik untuk hal-hal yang biasanya kita lihat di dunia sekitar kita, tetapi pada hal-hal "kuantum" (seperti qubit), pendekatan ini gagal, dan kita perlu memperbaruinya untuk menjelaskan perilaku yang kita lihat.

Jadi, apa solusinya? Untuk menggambarkan mekanika kuantum, kita dapat menggunakan *amplitudo probabilitas.* Amplitudo probabilitas mirip dengan probabilitas normal dalam hal:

- amplitudo memiliki nilai besaran,
- setiap hasil yang mungkin terjadi memiliki amplitudo probabilitas,
- dan besarnya amplitudo hasil itu memberi tahu kita seberapa besar kemungkinan hasil itu terjadi.

Tetapi amplitudo juga memiliki sifat tambahan yang kita sebut " [fase](gloss:phase) ". Anda dapat menganggap fase sebagai sudut, jadi jika peluang konvensional hanya memiliki besar, amplitudo juga memiliki arah.

![perbandingan probabilitas vs amplitudo](images/what-is/amp-vs-prob.svg)

Lebih spesifik lagi, amplitudo adalah bilangan kompleks (jika Anda tidak tahu apa itu bilangan kompleks, jangan khawatir tentang itu dulu).

Hasil dari fase adalah, ketika kita menambahkan dua amplitudo ini bersama-sama, mereka dapat saling meniadakan, seperti halnya bilangan positif dan negatif. Perilaku ini disebut *interferensi* dan menjelaskan semua perilaku khusus mekanika kuantum yang tidak kita lihat dalam mekanika klasik.

### Menjelaskan apa itu gerbang-H

<!-- ::: q-carousel -->

<!-- ::: div -->

![simbol untuk gerbang-H](images/what-is/carousel/0/0.png)

Jadi, dengan alat baru ini di kotak kita, dapatkah kita menggambarkan H-gate?

<!-- ::: -->


<!-- ::: div -->


![amplitudo awal](images/what-is/carousel/0/1.svg)

Melihat kami telah membawa Anda ke rute ini, Anda mungkin bisa menebak jawabannya adalah "ya", dan ini melibatkan fase. Mari kita mulai dengan mengingat bagaimana gerbang-H bekerja pada keadaan $|0\rangle$. Sebelum gerbang-H, peluang untuk mengukur keadaan $|0\rangle$ adalah 1, jadi besarnya amplitudo harus 1 juga. Kita tidak dapat mengetahui fase apa dari informasi ini, jadi kita hanya akan menganggapnya 0.

<!-- ::: -->

<!-- ::: div -->

![setelah penerapan 1 gerbang H](images/what-is/carousel/0/2.svg)

Setelah kita menerapkan gerbang-H, kita akan mengukur $|0\rangle$ atau $|1\rangle$ dengan probabilitas $1/2$. Karena kita mengkuadratkan besaran amplitudo untuk mendapatkan probabilitas, besaran kedua amplitudo ini harus $\sqrt{1/2}$. Kita juga dapat mengatakan hal yang sama jika kita telah memulai dalam keadaan $|1\rangle$. Sekali lagi, kita belum bisa mendapatkan informasi tentang bagaimana transformasi ini dapat mempengaruhi fase; mereka sebenarnya dapat diubah dengan cara apa pun dan kita tidak akan menyadarinya melalui eksperimen ini saja, jadi kita belum akan membuat tebakan apa pun.

<!-- ::: -->

<!-- ::: div -->

![setelah penerapan 2 gerbang-H (tanpa fase)](images/what-is/carousel/0/3.svg)

Tetapi ketika kita menerapkan dua gerbang-H, kita dapat mulai membuat pernyataan tentang fase-fase yang dapat dimiliki oleh amplitudo kita. Pada akhir pohon amplitudo, kita memiliki empat cabang berbeda yang hanya mengarah pada dua hasil yang mungkin. Untuk menghitung amplitudo akhir untuk hasil ini, kita perlu mengalikan di sepanjang cabang-cabang itu dan menambahkan amplitudo bersama-sama, seperti yang kita lakukan pada pohon probabilitas. Untuk memberikan peluang 0% untuk mengukur $|1\rangle$, kita memerlukan amplitudo pada dua cabang yang mengarah ke $|1\rangle$ untuk memiliki fase yang berlawanan.

<!-- ::: -->

<!-- ::: div -->

![bagaimana kita berpikir gerbang-H berperilaku pada kedua status input](images/what-is/carousel/0/4.svg)

Karena dua gerbang-H secara berurutan memberikan hasil yang deterministik, gerbang-H harus selalu berperilaku dengan cara yang sama jika diberikan input yang sama. Berikut ini satu tebakan untuk apa yang mungkin terjadi: Bertindak pada keadaan $|0\rangle$, semua amplitudo keluaran menunjuk ke arah yang sama, tetapi bekerja pada keadaan $|1\rangle$, fase dari $|1\rangle$ keadaan diputar 180°.

<!-- ::: -->

<!-- ::: div -->

![Mengikat pohon bersama-sama untuk memprediksi perilaku dua gerbang-H](images/what-is/carousel/0/5.svg)

Ketika kita mengikat pohon-pohon ini bersama-sama, kita mengalikan sepanjang cabang dan menjumlahkan amplitudo akhir untuk setiap keadaan.

<!-- ::: -->

<!-- ::: div -->

![Mengerjakan amplitudo akhir keadaan |0>](images/what-is/carousel/0/6.svg)

Mari kita mulai dengan dua cabang yang mengarah ke status $|0\rangle$. Untuk cabang teratas, kita melakukan $\sqrt{1/2}\times\sqrt{1/2} = 1/2$, dan kita menambahkannya ke cabang lain yang mengarah ke $|0\rangle$ (yang kita juga mendapatkan $1/2$). Jika dijumlahkan, magnitudo akhir dari amplitudo keadaan $|0\rangle$ adalah $\tfrac{1}{2} + \tfrac{1}{2} = 1$. Ini berarti probabilitas mengukur $|0\rangle$ juga $1$.

<!-- ::: -->

<!-- ::: div -->

![Mengerjakan amplitudo akhir keadaan |1>](images/what-is/carousel/0/7.svg)

Model kita bekerja sampai sejauh ini, mari kita lihat apa yang terjadi pada cabang yang mengarah ke status $|1\rangle$. Kedua cabang ini juga berakhir dengan besaran $1/2$, tetapi cabang di bawah menunjuk ke arah lain. Ini berarti dua amplitudo ini dibatalkan! Pada pohon amplitudo, kita telah menggunakan tanda minus untuk menunjukkan bahwa gerbang-H telah membalikkan arah amplitudo ini. Amplitudo akhir dari keadaan $|1\rangle$ adalah $0$, yang cocok dengan apa yang kita pikir akan terjadi!

<!-- ::: -->

<!-- ::: div -->

![Perilaku lengkap dari gerbang-H](images/what-is/carousel/0/4.svg)

Jadi model amplitudo kita berfungsi! Kita sekarang memiliki deskripsi lengkap tentang gerbang-H.

Di sini, sesuatu yang menarik dan kontra-intuitif telah terjadi; kita menggunakan probabilitas untuk menjelaskan kurangnya pengetahuan, misalnya jika kita melempar koin, itu bisa menjadi kepala atau ekor dengan probabilitas yang sama. Kenyataannya adalah bahwa koin akan berada di salah satu negara bagian ini dan kita menggunakan probabilitas untuk menghadapi fakta bahwa kita tidak tahu yang mana.

Tetapi dengan model amplitudo ini, kita tidak bisa mengatakan qubit mengambil rute tertentu (“$0 \rightarrow 0 \rightarrow 0$ *or* $0 \rightarrow 1 \rightarrow 0$”) karena kita tidak akan melihat efek interferensi. Ini membuat orang mengatakan hal-hal seperti "qubit bisa menjadi $0$ *dan* $1$ *pada saat yang sama!"*, yang belum tentu cara yang salah untuk menggambarkan perilaku qubit, tetapi juga tidak terlalu berguna bagi orang yang belajar tentang komputasi kuantum. Penting untuk diingat bahwa, karena kita tidak melihat perilaku seperti ini dalam kehidupan kita sehari-hari, kita tidak memiliki kata yang umum untuk itu, itulah sebabnya para ilmuwan menciptakan kata-kata baru yang akan kita pelajari dalam kursus ini.

## Merapikan apa yang sudah kita pelajari

Kita sekarang memiliki penjelasan bagaimana cara kerja H-gate menggunakan pohon amplitudo, tetapi menggunakan pohon amplitudo bisa menjadi berantakan dengan cepat. Ada begitu banyak cabang dan label sehingga bahkan pohon untuk sejumlah kecil qubit menjadi sangat rumit. Untuk menyederhanakan ini, ilmuwan komputer kuantum menggunakan objek matematika yang disebut ' [vektor](gloss:vector) ' yang, untuk tujuan kita, hanyalah daftar angka. Pada setiap titik dalam perhitungan kita, kita dapat menggunakan vektor untuk melacak amplitudo dari semua hasil yang mungkin.

![Gambar yang menunjukkan bagaimana keadaan komputer kuantum dapat direpresentasikan melalui daftar amplitudo pada titik mana pun dalam komputasi](images/what-is/amp-vectors.svg)

Kita sekarang memiliki cara yang cukup sederhana untuk mensimulasikan qubit, jadi apa gunanya mencoba membuat komputer kuantum yang berfungsi ketika kita sudah tahu cara mensimulasikannya dengan sempurna menggunakan komputer tradisional? Nah, untuk sedikit qubit, simulasi itu mudah, tetapi jika kita ingin mensimulasikan komputer kuantum dengan banyak qubit, melacak semua amplitudo ini bisa sangat memakan waktu dan membutuhkan banyak memori.

Jumlah hasil yang mungkin berlipat ganda dengan setiap qubit ekstra yang kita tambahkan, dan jika kita menggunakan teknik ini, ukuran vektor kita akan tumbuh secara eksponensial juga. Ini tidak berlaku untuk setiap sirkuit kuantum, misalnya, jika kita memulai dengan semua qubit kita dalam keadaan 0 dan tidak melakukan apa pun pada mereka, maka cukup mudah untuk memprediksi seperti apa outputnya. Tetapi tampaknya sirkuit kuantum yang lebih sulit hanya dapat disimulasikan pada komputer klasik melalui algoritma yang tumbuh secara eksponensial dengan jumlah qubit di sirkuit. Batas atas untuk mensimulasikan sirkuit kuantum yang sulit cenderung berada di antara tanda 30-40 qubit.

## Ringkasan

Kita telah belajar banyak dalam bab ini, tetapi poin-poin utamanya adalah:

- Kita dapat membuat bit dari objek yang mengikuti aturan mekanika kuantum, dan kita menyebutnya "qubit"
- Qubit ini dapat dijelaskan menggunakan amplitudo probabilitas, yang seperti probabilitas klasik tetapi dengan "fase"
- Amplitudo ini dapat membatalkan satu sama lain (efek yang disebut interferensi) dan inilah yang menyebabkan perilaku yang sebelumnya tidak dapat dijelaskan
- Algoritma terbaik yang kita miliki untuk mensimulasikan qubit membutuhkan sumber daya yang tumbuh secara eksponensial dengan jumlah qubit, jadi mensimulasikan qubit dalam jumlah besar tidak dapat dilakukan oleh komputer klasik.

Kita telah melihat bahwa komputer kuantum dapat melakukan sesuatu yang tidak dapat dilakukan komputer klasik: mensimulasikan perilaku qubit! Komputer klasik membutuhkan sumber daya eksponensial, tetapi komputer kuantum hanya dapat melakukan operasi dan melihat apa yang terjadi. Pertanyaan menarik berikutnya adalah: “Dapatkah komputer kuantum memecahkan masalah lain yang tidak bisa dilakukan komputer klasik?”. Dalam kursus ini kita akan melihat jawabannya adalah "ya"!
