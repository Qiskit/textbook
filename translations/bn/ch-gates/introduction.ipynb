# ভূমিকা

আমরা একক কিউবিট এবং একক কিউবিট গেটগুলির সাথে কিছু আকর্ষণীয় প্রভাব দেখেছি, তবে কোয়ান্টাম কম্পিউটিংয়ের প্রকৃত শক্তি কিউবিটগুলির মধ্যে মিথস্ক্রিয়াগুলির মাধ্যমে উপলব্ধি করা হয়। এই বিভাগে আমরা একাধিক কিউবিট গেট প্রবর্তন করব এবং মাল্টি-কুবিট সিস্টেমের আকর্ষণীয় আচরণগুলি অন্বেষণ করব।

সাধারণত, হার্ডওয়্যারে সরাসরি প্রয়োগ করা যেতে পারে এমন গেটগুলি শুধুমাত্র এক বা দুটি কিউবিটে কাজ করবে। আমাদের সার্কিটগুলিতে, আমরা জটিল গেটগুলি ব্যবহার করতে চাই যা প্রচুর সংখ্যক কিউবিটগুলিতে কাজ করে। সৌভাগ্যবশত, এই একটি সমস্যা হবে না. হার্ডওয়্যার দ্বারা আমাদের দেওয়া এক এবং দুটি কিউবিট গেট দিয়ে, অন্য কোনও গেট তৈরি করা সম্ভব।

এই অধ্যায়ে আমরা প্রথমে সবচেয়ে মৌলিক মাল্টি-কুবিট গেটগুলির সাথে পরিচয় করিয়ে দেব, সেইসাথে তাদের বর্ণনা ও বিশ্লেষণ করতে ব্যবহৃত গণিত। তারপরে আমরা দেখাব কিভাবে প্রমাণ করা যায় যে এই গেটগুলি যেকোনো সম্ভাব্য কোয়ান্টাম অ্যালগরিদম তৈরি করতে ব্যবহার করা যেতে পারে। তারপর অধ্যায়টি কোয়ান্টাম গেটগুলির ছোট আকারের ব্যবহারগুলি দেখে শেষ হয়। উদাহরণস্বরূপ, আমরা দেখি কিভাবে একক- এবং দুই-কুবিট অপারেশন থেকে টফোলির মতো তিন-কুবিট গেট তৈরি করা যায়।
