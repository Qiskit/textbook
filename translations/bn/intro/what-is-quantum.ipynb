# কোয়ান্টাম কি?

## বিট কি?

আগের অধ্যায়ে আমরা শিখেছি যে 'বিট' হল বিশ্বের সবচেয়ে সহজ একক, এবং যে কোনো তথ্য উপস্থাপনের জন্য এটি ব্যবহার করা যেতে পারে। আমরা আরও শিখেছি যে সমস্ত আধুনিক কম্পিউটার বিটের মাধ্যমে কাজকর্ম সঞ্চয় এবং সম্পন্ন করে। এখন পর্যন্ত, আমরা বিটগুলোকে তথ্যের বিমূর্ত একক হিসাবে চিন্তা করেছি, তবে একটি কার্যকরী কম্পিউটার তৈরি করতে, বিটগুলোকে আমাদের বাস্তব জিনিস থেকে তৈরি করতে হবে। তাহলে আমরা বাস্তব জগতে কিভাবে একটি বিট তৈরি করতে পারি?

<!-- ::: tab -->

### পাঞ্চড কার্ড

<!-- ::: column(width=350) -->

**পাঞ্চড কার্ড**

কম্পিউটিংয়ের প্রথম দিকে, কম্পিউটার বিজ্ঞানীরা কাগজের কার্ডে ছিদ্র করে বিট সংরক্ষণ করতেন। এই কার্ডগুলিকে গ্রিডে বিভক্ত করা হয়েছিল, এবং গ্রিডের প্রতিটি কক্ষ কিছুটা প্রতিনিধিত্ব করে। সেই কক্ষে একটি ছিদ্র থাকলে, বিটটি '1', যদি কোনো ছিদ্র না থাকে তবে বিটটি '0'। আপনি যদি একটি বিট মান পরিবর্তন করতে চান, আপনি হয় একটি নতুন গর্ত ঘুষি, অথবা গর্ত ব্যাক আপ [প্যাচ](gloss:patch) করতে পারেন.

### কমপ্যাক্ট ডিস্ক

<!-- ::: column(width=350) -->

**কমপ্যাক্ট ডিস্ক**

80'র দশকে বিট ব্যবহার করে অডিও রেকর্ডিং সংরক্ষণ করার জন্য কমপ্যাক্ট ডিস্ক তৈরি করা হয়েছিল। ডিজাইনাররা বিটগুলোকে বোঝানোর জন্য ডিস্কের পৃষ্ঠে গর্ত এবং কুন্ড ব্যবহার করার সিদ্ধান্ত নিয়েছিলেন এবং আমরা সে বিটগুলোকে সনাক্ত করতে ডিস্কের উপর একটি লেজার ছুড়ে দিই। ডিস্ক একটি সর্পিল লাইন বরাবর পড়া হয়, যা <br>অনেক অংশে বিভক্ত। প্রতিটি অংশের জন্য, একটি সমতল পৃষ্ঠ একটি '0' বোঝায় এবং চূড়া থেকে খাদে (বা এর বিপরীত) একটি '1' বোঝায়।

### ইলেকট্রন অরবিটাল

<!-- ::: column(width=350) -->

**ইলেকট্রন অরবিটাল**

আপনি রসায়ন থেকে মনে রাখতে পারেন (না পারলে চিন্তা নাই) যে ইলেকট্রন নামক কণাগুলি একটি পরমাণুর কেন্দ্রকে (নিউক্লিয়াস) প্রদক্ষিণ করে। আপনি হয়তো এটাও মনে রাখতে পারেন যে এই ইলেকট্রনগুলি শুধুমাত্র নিউক্লিয়াস থেকে নির্দিষ্ট ( [কয়ান্টাইজড](gloss:quantized) ) দূরত্বে প্রদক্ষিণ করতে পারে যাকে আমরা [*শেল *](gloss:shell) বলি। ইলেক্ট্রন কোন শেলে অবস্থিত আছে তা শনাক্ত করার উপায় যদি আমাদের কাছে তাহলে আমরা এই দুটি শেল বাছাই করতে পারি এবং ইলেক্ট্রনের অবস্থানকে বিট হিসেবে ব্যবহার করতে পারি। ইলেকট্রন যদি এই শেলগুলির একটিতে থাকে তবে বিটটি '0', আর যদি অন্যটিতে থাকে তবে বিটটি '1'।

*ক্লাসিকাল পদার্থবিদ্যা* ব্যবহার করে কার্ড এবং কমপ্যাক্ট ডিস্ক উভয়ের আচরণই সহজে ব্যাখ্যা করা যেতে পারে: পদার্থবিদ্যা হল মূলত আমাদের চারপাশের বিশ্বের একটি বর্ণনা; পৃথিবী কেমন আচরণ করে তা দেখার জন্য আমরা পরীক্ষা-নিরীক্ষা করি এবং এ থেকে মহাবিশ্ব কী নিয়ম অনুসরণ করে আমরা তা খুঁজে বের করার চেষ্টা করি। "ক্লাসিকাল পদার্থবিদ্যা" নামটি হল কিছু সংখ্যক নিয়মের সেট, যা বিজ্ঞানীরা 1900' সালের আগে নিয়ে এসেছিলেন এবং এটি ক্রিকেট বল বা গাড়ির ইঞ্জিনের মতো জিনিসগুলির আচরণের ভবিষ্যদ্বাণী করতে বেশি ভাল।

কিন্তু পরমাণুকে প্রদক্ষিণ করা ইলেকট্রন একটু ভিন্ন। 1900 এর দশকের গোড়ার দিকে বিজ্ঞানীরা এটমিক স্কেলে জিনিসগুলি অধ্যয়ন করতে সক্ষম হন। তারা দেখতে পান যে, পরমাণুর মতো ক্ষুদ্র জিনিসগুলো আমাদের দৈনন্দিন জীবনে যে জিনিসগুলোর সাথে আমরা অভ্যস্থ, তার থেকে ভিন্নভাবে আচরণ করে এবং কিছু ক্ষেত্রে ক্লাসিকাল পদার্থবিজ্ঞানের নিয়মগুলি পুরোপুরি সঠিক ছিল না। তাই পদার্থবিদ্যার পরিবর্তনের যে প্রয়োজন ছিল, তাতে বিজ্ঞানীরা আরও সঠিক নিয়ম নিয়ে এসেছিলেন যা "কোয়ান্টাম পদার্থবিদ্যা" নামে পরিচিত।

### কোয়ান্টাম পদার্থবিদ্যার মূল পরীক্ষাসমূহ

আগ্রহীদের জন্য, এখানে কিছু ইভেন্টের ছোট্ট সারসংক্ষেপসমূহ রয়েছে যা কোয়ান্টাম পদার্থবিজ্ঞান বিকশিত হতে পরিচালিত করেছিল। এই কোর্সের বাকি অংশ পড়ার জন্য আপনাকে এইগুলি বোঝার দরকার নেই। যারা ইতিহাস এবং পদার্থবিদ্যা সম্পর্কে আরও পড়তে চান তাদের জন্যই কেবল এটি।

<summary>অতিবেগুনী বিপর্যয়</summary><p> গরম জিনিসগুলি আলো হিসাবে বিকিরণ নির্গত করে এবং তাপমাত্রা যত বেশি হবে, এই আলোক তরঙ্গগুলির ফ্রিকোয়েন্সি তত বেশি হবে। যদি কিছু শুধুমাত্র আলো নির্গত এবং বিকিরণ করে (অর্থাৎ, এটি প্রতিফলিত করে না), আমরা তাকে 'কালো দেহ' বলি। তাত্ত্বিকভাবে নিখুঁত কালো শরীর নক্ষত্র এবং গ্রহ সহ অনেক বাস্তব বস্তুর একটি ভয়ানক অনুমান নয়।</p>
<p> 1900-1905 সালের মধ্যে, দুই বিজ্ঞানী (Rayleigh এবং Jeans) শাস্ত্রীয় পদার্থবিজ্ঞানের নিয়মগুলি ব্যবহার করে একটি সমীকরণ তৈরি করেছিলেন যা বর্ণনা করে যে কীভাবে একটি কালো দেহের <i>তাপমাত্রা</i> তার নির্গত আলোক তরঙ্গের <i>ফ্রিকোয়েন্সিকে</i> প্রভাবিত করে। তাদের উদ্ভূত আইনটি ভবিষ্যদ্বাণী করেছিল যে গরম বস্তুগুলি হাস্যকর পরিমাণে উচ্চ-ফ্রিকোয়েন্সি (আল্ট্রাভায়োলেট) আলো নির্গত করবে, যা আমরা যা দেখি তার সাথে মেলে না।</p>
<p> এই বিন্দু পর্যন্ত, এটি ব্যাপকভাবে গৃহীত হয়েছিল যে আলো একটি তরঙ্গ। কিন্তু আরেকজন বিজ্ঞানী (ম্যাক্স প্ল্যাঙ্ক) দেখাতে পেরেছেন যে, যদি আলোকে পরিমাপকৃত অংশে নির্গত করতে হয় (অর্থাৎ, কণাকে আমরা এখন 'ফোটন' বলি), আপনি এমন একটি সমীকরণ বের করতে পারেন যা তাপমাত্রা এবং নির্গমনের মধ্যে সম্পর্ককে সঠিকভাবে বর্ণনা করতে পারে যা আমরা করতে পারি। পরিমাপ করা.</p>

<summary>ডবল স্লিট পরীক্ষা</summary><p> আপনি যদি কিছু কণাকে একটি ছিদ্র দিয়ে একটি স্ক্রিনে ফায়ার করেন এবং যেখানে তারা অবতরণ করেন তা রেকর্ড করেন, আপনি স্ক্রিনের মাঝখানে আরও বেশি এবং প্রান্তে কম দেখতে পাবেন (আকৃতিটি আসলে এর চেয়ে কিছুটা জটিল তবে আমরা আলোচনা করব না যে এখানে)। তাহলে, আপনি যদি পর্দায় _দুটি_ ছিদ্রের মাধ্যমে কণা ছুঁড়ে ফেলেন তাহলে আপনি কী ঘটতে আশা করবেন? যদি গর্তগুলি যথেষ্ট কাছাকাছি হত, সাধারণ জ্ঞান বলে যে আমরা মোটামুটি একই জিনিস দেখতে চাই, সম্ভবত আরও কণা এটি দিয়ে তৈরি করে।</p>
<p> কিন্তু যখন আমরা ফোটন, ইলেকট্রন এবং এমনকি পরমাণুর মতো ছোট কণার সাথে এই পরীক্ষাটি করার চেষ্টা করি, তখন আমরা একটি হস্তক্ষেপ প্যাটার্ন দেখতে পাই, এমনকি যদি আমরা একবারে একটি কণা প্রেরণ করি! এই পরীক্ষাটি 1700 এর দশকের শেষের দিকে আলোর সাথে করা হয়েছিল, যা বিজ্ঞানীদের বিশ্বাস করে যে আলো একটি কণা নয় বরং একটি তরঙ্গ ছিল। কিন্তু আমরা যেমন অতিবেগুনী বিপর্যয়ের সাথে দেখেছি, এটি এখনও সম্পূর্ণ চিত্র ছিল না।</p> ![](images/what-is/Single-slit-and-double-slit.jpg)<p> Jordgette দ্বারা ছবি - নিজের কাজ, CC BY-SA 3.0, <a>লিঙ্ক</a></p>

কম্পিউটার বিজ্ঞানীদের জন্য, একটি স্বাভাবিক প্রশ্ন হল: "কি হবে যদি আমাদের বিটগুলি ক্লাসিক্যাল পদার্থবিজ্ঞানের পরিবর্তে কোয়ান্টাম পদার্থবিজ্ঞানের নিয়মগুলি অনুসরণ করে?" আমরা এই বিটগুলিকে "কোয়ান্টাম বিট" এর জন্য "কিউবিট" বলি, এবং যে কম্পিউটারগুলি এই বিটগুলিতে কাজ করতে পারে তাদের "কোয়ান্টাম কম্পিউটার" বলা হয়।

পরবর্তী সেকশনগুলোয় আপনি এই কিউবিটগুলির মধ্যে একটির সাথে ইন্টারঅ্যাক্ট করতে পারেন (সত্যি সত্যি না - আমরা কেবল এটা সিমুলেট করব - তবে আমরা প্রতিশ্রুতি দিচ্ছি যে আচরণ একই হবে) এবং 'কোয়ান্টাম' কী তা আবিষ্কার করতে পারেন!

## কিউবিট পরিচিতি

কোয়ান্টাম আচরণ ঘাটাঘাটি করতে, আমাদের নিজেদেরকে প্রবাবিলিটি সম্পর্কে মনে করিয়ে দিতে হবে। আপনি যদি পূর্ববর্তী অধ্যায়টি পড়ে থাকেন, আমরা এখন পর্যন্ত যে সমস্ত অপারেশন দেখেছি তা [নির্ধারণমূলক](gloss:deterministic) । এর মানে হল, একই ইনপুট স্টেটে কাজ করলে, তারা সবসময় একই আউটপুট স্টেট দেবে। আসুন একটি নতুন অপারেশন দেখে নেওয়া যাক যা শুধুমাত্র কোয়ান্টাম বিটগুলিতে কাজ করে: [হাদামার্ড](gloss:hadamard) গেট, যাকে আমরা সংক্ষেপে "H-gate" বলব।

আমরা এইমাত্র কিছু আকর্ষণীয় আচরণ দেখেছি: দুটি সম্ভাব্য ক্রিয়াকলাপ, ক্রমানুসারে প্রয়োগ করা, একে অপরকে "পূর্বাবস্থায় ফেরানো" বলে মনে হচ্ছে! আসলে, এই সাধারণ আচরণটি এতটাই অস্বাভাবিক যে আমরা সম্ভাব্যতা গাছ ব্যবহার করে এটি বর্ণনা করতে পারি না। এমন কোন সংখ্যা নেই যা আমরা আমাদের শাখাগুলিতে রাখতে পারি যা সমস্ত ক্ষেত্রে সঠিক ফলাফল দেয়।

আপনি হয়তো ভাবছেন, "হয়তো H-গেট জানে যে এটি দুবার প্রয়োগ করা হচ্ছে, এবং ভিন্নভাবে আচরণ করে?" এটি কিউবিটের অবস্থার পাশাপাশি গেটে আরেকটি ইনপুট হিসাবে গণনা করবে। আপনি এই আচরণটি বর্ণনা করতে পারেন যদি আপনি H-গেট অপারেশনে অন্য লুকানো ইনপুটকে অনুমতি দেন যা আমরা সরাসরি দেখতে পাই না। এবং আসলে, এই কি ঘটছে সাজানোর.

## সম্ভাবনার বাইরে

প্রবাবিলিটীগুলি কার্যকর হয় যখন অনেকগুলি সম্ভাব্য ফলাফল থাকে এবং আমাদের কাছে কাজ করার জন্য পর্যাপ্ত তথ্য নেই যা ঘটবে, যেমন একটি ডাইস রোল, বা একটি মুদ্রা টস৷ আমরা প্রতিটি ফলাফলকে একটি সম্ভাব্যতা দিই এবং কিছু ঘটার সম্ভাবনা খুঁজে বের করতে এগুলি ব্যবহার করি। আমরা সাধারণত আমাদের চারপাশের বিশ্বে যে জিনিসগুলি দেখি তার জন্য সম্ভাবনাগুলি অত্যন্ত ভাল কাজ করে, কিন্তু "কোয়ান্টাম" জিনিসগুলির সাথে (যেমন কিউবিট), এই পদ্ধতিটি ব্যর্থ হয় এবং আমরা যে আচরণটি দেখছি তা ব্যাখ্যা করার জন্য আমাদের এটি আপডেট করতে হবে।

তাহলে, সমাধান কি? কোয়ান্টাম মেকানিক্স বর্ণনা করতে, আমরা প্রবাবিলিটি *এমপ্লিটিউড*  ব্যবহার করতে পারি। প্রবাবিলিটি এমপ্লিটিউড সাধারণ প্রবাবিলিটি মতো:

- এমপ্লিটিউডের মাত্রা অথবা মান আছে,
- প্রতিটি সম্ভাব্য ফলাফলের একটি প্রবাবিলিটি এমপ্লিটিউড আছে,
- এবং সেই ফলাফলের এমপ্লিটিউড মাত্রা আমাদের বলে যে ফলাফলটি ঘটার সম্ভাবনা কতটা।

কিন্তু এমপ্লিটিউডগুলো একটি অতিরিক্ত বৈশিষ্ট্য রয়েছে যাকে আমরা " [phase](gloss:phase) " বলি। আপনি ফেজকে একটি কোণ হিসাবে ভাবতে পারেন, তাই যেখানে প্রচলিত সম্ভাবনার শুধুমাত্র একটি মাত্রা থাকে।এমপ্লিটিউড এরও একটি দিক থাকে।

![comparison of probabilities vs amplitudes](images/what-is/amp-vs-prob.svg)

আরও নির্দিষ্টভাবে, একটি প্রশস্ততা একটি জটিল সংখ্যা (যদি আপনি একটি জটিল সংখ্যা কী তা না জানেন তবে এখনও এটি সম্পর্কে চিন্তা করবেন না)।

ফেজের ফলাফল হল, যখন আমরা এই দুটি প্রশস্ততা একসাথে যোগ করি, তখন তারা একে অপরকে বাতিল করতে পারে, ঠিক যেমন ধনাত্মক এবং ঋণাত্মক সংখ্যাগুলি করে। এই আচরণটিকে *হস্তক্ষেপ* বলা হয় এবং কোয়ান্টাম মেকানিক্সের জন্য নির্দিষ্ট সমস্ত আচরণ ব্যাখ্যা করে যা আমরা ক্লাসিক্যাল মেকানিক্সে দেখি না।

যখন আমরা এই গাছগুলিকে একত্রে শৃঙ্খলিত করি, তখন আমরা শাখা বরাবর সংখ্যাবৃদ্ধি করি এবং প্রতিটি রাজ্যের জন্য চূড়ান্ত প্রশস্ততা যোগ করি।

<!-- ::: -->

<!-- ::: div -->

![Working out the final amplitude of the state |0>](images/what-is/carousel/0/6.svg)

স্টেট $|0\rangle$ এর দিকে নিয়ে যাওয়া দুটি শাখা দিয়ে শুরু করা যাক। শীর্ষ শাখার জন্য, আমরা $\sqrt{1/2}\times\sqrt{1/2} = 1/2$ করি, এবং আমরা এটিকে অন্য শাখায় যোগ করি যা $|0\rangle$ (যার জন্য আমরা $1/2$ও পাই)। এগুলোকে একসাথে যোগ করলে, $|0\rangle$ স্টেটের প্রশস্ততার চূড়ান্ত মাত্রা হল $\tfrac{1}{2} + \tfrac{1}{2} = 1$। এর মানে হল $|0\rangle$ পরিমাপের সম্ভাবনাও $1$।

<!-- ::: div -->

![Working out the final amplitude of the state |1>](images/what-is/carousel/0/7.svg)

আমাদের মডেল এখন পর্যন্ত কাজ করে, দেখা যাক যে শাখাগুলি $|1\rangle$ রাজ্যে নিয়ে যায় সেগুলির কী হয়৷ এই উভয় শাখাই $1/2$ এর মাত্রায় শেষ হয়, কিন্তু নীচের শাখাটি অন্য দিকে নির্দেশ করছে। এর মানে এই দুটি প্রশস্ততা বাতিল! প্রশস্ততা গাছে, আমরা দেখানোর জন্য একটি বিয়োগ চিহ্ন ব্যবহার করেছি যে এইচ-গেট এই প্রশস্ততার দিকটিকে বিপরীত করেছে। রাষ্ট্রের চূড়ান্ত প্রশস্ততা $|1\rangle$ তারপর $0$, যা আমরা যা ভেবেছিলাম তার সাথে মিলে যায়!

তাই আমাদের প্রশস্ততা মডেল কাজ করে! আমাদের কাছে এখন এইচ-গেটের সম্পূর্ণ বিবরণ আছে।

এখানে, আকর্ষণীয় এবং পাল্টা স্বজ্ঞাত কিছু ঘটেছে; আমরা জ্ঞানের অভাবের জন্য সম্ভাব্যতা ব্যবহার করি, যেমন আমরা যদি একটি মুদ্রা টস করি, তবে এটি সমান সম্ভাবনার সাথে মাথা বা লেজ হতে পারে। বাস্তবতা হল যে মুদ্রাটি এই রাজ্যগুলির মধ্যে একটিতে থাকবে এবং আমরা সম্ভাব্যতাগুলি ব্যবহার করি এই সত্যটি মোকাবেলা করার জন্য যে আমরা জানি না কোনটি।

কিন্তু এই প্রশস্ততা মডেলের সাথে, আমরা বলতে পারি না যে কিউবিট একটি নির্দিষ্ট রুট নিয়েছে (“$0 \rightarrow 0 \rightarrow 0$ *বা* $0 \rightarrow 1 \rightarrow 0$”) কারণ আমরা হস্তক্ষেপের প্রভাব দেখতে পাব না। এটি লোকেদের "কুবিট *একই সময়ে* $0$ *এবং* $1$ হতে পারে!" এর মত কিছু বলতে চালিত করে। , যা অগত্যা কিউবিটের আচরণ বর্ণনা করার জন্য একটি ভুল উপায় নয়, তবে কোয়ান্টাম কম্পিউটিং সম্পর্কে শেখা লোকেদের জন্য বিশেষভাবে কার্যকর নয়। এটা মনে রাখা গুরুত্বপূর্ণ, কারণ আমরা আমাদের দৈনন্দিন জীবনে এই ধরনের আচরণ দেখতে পাই না, আমাদের কাছে এটির জন্য সত্যিই সাধারণ শব্দ নেই, এই কারণেই বিজ্ঞানীরা নতুন শব্দ তৈরি করেছেন যা আমরা এই কোর্সে শিখব।

## পরিশেষে,

আমাদের কাছে এখন প্রশস্ততা গাছ ব্যবহার করে H-gate এর একটি কার্যকরী ব্যাখ্যা রয়েছে, তবে গাছ ব্যবহার করা খুব দ্রুত অগোছালো হতে পারে। এমন অনেকগুলি শাখা এবং লেবেল রয়েছে যে এমনকি অল্প সংখ্যক কিউবিটের জন্য গাছগুলিও খুব জটিল হয়ে যায়। এটিকে সহজ করার জন্য, কোয়ান্টাম কম্পিউটার বিজ্ঞানীরা ' [ভেক্টর](gloss:vector) ' নামক গাণিতিক বস্তু ব্যবহার করেন যা আমাদের উদ্দেশ্যে, সংখ্যার তালিকা মাত্র। আমাদের গণনার প্রতিটি পয়েন্টে, আমরা সম্ভাব্য সমস্ত ফলাফলের প্রশস্ততার ট্র্যাক রাখতে একটি ভেক্টর ব্যবহার করতে পারি।

![Image showing how the state of a quantum computer can be represented through a list of amplitudes at any point in the computation](images/what-is/amp-vectors.svg)

আমাদের কাছে এখন কিউবিট সিমুলেট করার একটি খুব সহজ উপায় রয়েছে, তাই একটি কার্যকরী কোয়ান্টাম কম্পিউটার তৈরি করার চেষ্টা করার অর্থ কী যখন আমরা ইতিমধ্যেই জানি কীভাবে এটিকে প্রচলিত কম্পিউটার ব্যবহার করে পুরোপুরি অনুকরণ করা যায়? ঠিক আছে, অল্প সংখ্যক কিউবিটের জন্য, সিমুলেট করা সহজ, তবে আমরা যদি প্রচুর সংখ্যক কিউবিট সহ কোয়ান্টাম কম্পিউটার সিমুলেট করতে চাই তবে এই সমস্ত প্রশস্ততার ট্র্যাক রাখা খুব সময়সাপেক্ষ এবং প্রচুর মেমরির প্রয়োজন হতে পারে।

## সারসংক্ষেপ

আমরা এই অধ্যায়ে অনেক কিছু শিখেছি, কিন্তু মূল বিষয়গুলো হল:

- আমরা কোয়ান্টাম মেকানিক্সের নিয়ম অনুসরণ করে এমন বস্তুগুলি থেকে বিট তৈরি করতে পারি এবং আমরা এইগুলিকে "কিউবিট" বলি।
- এই কিউবিটগুলিকে সম্ভাব্যতা প্রশস্ততা ব্যবহার করে বর্ণনা করা যেতে পারে, যা ক্লাসিক্যাল সম্ভাব্যতার মতো কিন্তু "ফেজ" সহ
- এই প্রশস্ততাগুলি একে অপরকে বাতিল করতে পারে (একটি প্রভাব যাকে হস্তক্ষেপ বলা হয়) এবং এটিই পূর্বের অব্যক্ত আচরণের কারণ হয়
- কিউবিট সিমুলেট করার জন্য আমাদের কাছে থাকা সেরা অ্যালগরিদমগুলি কিউবিটের সংখ্যা সহ সূচকীয় সংস্থানগুলি ব্যবহার করে, তাই প্রচুর সংখ্যক কিউবিট সিমুলেট করা ক্লাসিক্যাল কম্পিউটারের নাগালের বাইরে।

আমরা দেখেছি যে কোয়ান্টাম কম্পিউটার এমন কিছু করতে পারে যা ক্লাসিক্যাল কম্পিউটার করতে পারে না: qubits এর আচরণের অনুকরণ! ধ্রুপদী কম্পিউটারের সূচকীয় সংস্থান প্রয়োজন, কিন্তু কোয়ান্টাম কম্পিউটার কেবল অপারেশন চালাতে পারে এবং দেখতে পারে কি হয়। পরবর্তী আকর্ষণীয় প্রশ্ন হল: "একটি কোয়ান্টাম কম্পিউটার কি অন্য কোন সমস্যার সমাধান করতে পারে যা একটি ক্লাসিক্যাল কম্পিউটার করতে পারে না?"। এই কোর্সে আমরা দেখব উত্তর হল “হ্যাঁ”!

## অনুশীলন

এই পৃষ্ঠার সমস্ত qubits এই প্রশস্ততা মডেল ব্যবহার করে ডিজাইন করা ক্লাসিক্যাল প্রোগ্রাম ব্যবহার করে সিমুলেট করা হয়েছে, তাহলে আপনি কেন বিশ্বাস করবেন যে প্রকৃত কিউবিটগুলি কীভাবে আচরণ করে? আপনি যদি সত্যিকারের কোয়ান্টাম কম্পিউটারে *বাস্তব* কিউবিটগুলির সাথে খেলতে চান তবে [IBM কোয়ান্টাম কম্পোজারটি দেখুন](https://quantum-computing.ibm.com/composer/files/new) । কেন আপনি চেষ্টা করবেন না এবং এই পরীক্ষাটি পুনরায় তৈরি করুন এবং এটি একটি ডিভাইসে চালান?

আপনি সেখানে অনেক অন্যান্য গেট এবং নির্দেশাবলী দেখতে পাবেন, যার বেশিরভাগই আমরা আগে কভার *করিনি* । IBM-এর কম্পোজারে, যদি একটি নির্দেশ ধূসর হয়, তাহলে এটি একটি কোয়ান্টাম গেট নয় এবং আমরা এই অধ্যায়ে যে প্রশস্ততার নিয়মগুলি আবিষ্কার করেছি তা অনুসরণ করবে না।
